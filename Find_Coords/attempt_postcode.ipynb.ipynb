```python
import pandas as pd
df_raw = pd.read_csv(r'Path', encoding='ISO-8859-1')
df_raw.head()
```

```python
df_raw['point_coord'] = df_raw['point_coord'].str.replace("POINT(", "")
df_raw['point_coord'] = df_raw['point_coord'].str.replace(")", "")
df_raw[["Lat", "Long"]] = df_raw["point_coord"].str.split(' ', expand = True)
df_raw.drop('point_coord', axis = 1, inplace = True)
df_raw['Long'] = df_raw['Long'].astype(float)
df_raw['Lat'] = df_raw['Lat'].astype(float)
df_raw.head()
```

```python
df_raw.to_csv('postcode_malaysia.csv')
```

Filter those who contains the parsed "Region"\
For example "KLCC" founded in two rows ("Kuala Lumpur City Centre (KLCC)" & "Persiaran KLCC")\
Add these Long Lat up and divide by 2

```python
for i in not_included:
    filtered = df[df["street"].str.contains(i)]
    cal_lat = round((filtered["Lat"].sum())/(filtered["Lat"].count()), 4)
    cal_long = round((filtered["Long"].sum())/(filtered["Long"].count()), 4)
    print(f"{i}: {cal_lat} {cal_long}")
```
